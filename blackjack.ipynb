```python
from random import choice
from IPython.display import clear_output

class Blackjack():
    suit = {'h': 'hearts', 'd': 'diamonds', 's':'spades', 'c':'clubs'}
    original_deck = ['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'h7', 'h8', 'h9', 'h10', 'h11', 'h12', 'h13',
                    'd1', 'd2', 'd3', 'd4', 'd5', 'd6', 'd7', 'd8', 'd9', 'd10', 'd11', 'd12', 'd13',
                    's1', 's2', 's3', 's4', 's5', 's6', 's7', 's8', 's9', 's10', 's11', 's12', 's13',
                    'c1', 'c2', 'c3', 'c4', 'c5', 'c6', 'c7', 'c8', 'c9', 'c10', 'c11', 'c12', 'c13'
                    ]
    
    def __init__(self, player_score = 0, dealer_score = 0, player_cards= {}, dealer_cards={}, player_delt = 1, dealer_delt = 1):
        self.deck = self.original_deck.copy()
        self.player_score = player_score
        self.dealer_score = dealer_score
        self.player_cards = player_cards
        self.dealer_cards = dealer_cards
        self.player_delt = player_delt
        self.dealer_delt = dealer_delt
        
#   Helper Functions      
    def pick_card(self):
        card = choice(self.deck)
        self.deck.remove(card)
        return card
    def show_cards(self, dictionary):
        for value in dictionary.values():
            print(f'{value[1:]} of {self.suit[value[0]]}')
        return ""
    def reset(self):
        self.player_score = 0
        self.dealer_score = 0
        self.player_cards = {}
        self.dealer_cards = {}
        self.player_delt = 1
        self.dealer_delt = 1
        self.deck = self.original_deck.copy()
    def clear(self):
        clear_output()
        
    def deal_player(self):
        if self.player_score == 0:
            while self.player_delt <= 2:
                card = self.pick_card()
                self.player_score += int(card[1:])
                self.player_cards[self.player_delt] = card
                print(f'Card {self.player_delt}: {card[1:]} of {self.suit[card[0]]}')
                self.player_delt += 1
            print(f'Your card count = {self.player_score}')
            if self.player_score == 21:
                print('Blackjack, you won!')
                self.reset()
            elif self.player_score > 21:
                print('Busted, you lose.')
                self.reset()
            else:
                self.deal_dealer()
        else:
            print('Cards have already been delt. Choose hit or stand.')
        
    def deal_dealer(self):
        while self.dealer_delt <= 2:
            card = self.pick_card()
            self.dealer_score += int(card[1:])
            self.dealer_cards[self.dealer_delt] = card
            if self.dealer_delt == 2:
                print("Dealer's face up card.")
                print(f'{card[1:]} of {self.suit[card[0]]}')
            self.dealer_delt += 1
        if self.dealer_score > 21:
            print('You won, dealer busts!')
            print(self.show_cards(self.dealer_cards))
            print(f"Dealer's score: {self.dealer_score}")
            self.reset()
    def hit(self):
        self.player_delt += 1 
        card = self.pick_card()
        self.player_score += int(card[1:])
        self.player_cards[self.player_delt] = card
        print(f'{card[1:]} of {self.suit[card[0]]}')
        print(f'Your card count = {self.player_score}')
        if self.player_score > 21:
            print('Busted, you lose.')
            self.reset()
        
    def stand(self):
        while self.dealer_score < 17:
            self.dealer_delt += 1
            card = self.pick_card()
            self.dealer_score += int(card[1:])
            self.dealer_cards[self.dealer_delt] = card
        if self.dealer_score > 21:
            print('You won, dealer busts!')
            print(self.show_cards(self.dealer_cards))
            print(f"Dealer's score: {self.dealer_score}")
            self.reset()
        else:
            self.compare_cards()
            
    def compare_cards(self):
        print("Dealer's cards")
        self.show_cards(self.dealer_cards)
        print(f"Dealers score: {self.dealer_score}")
        print("Your cards")
        self.show_cards(self.player_cards)
        print(f"Your score: {self.player_score}")
        if self.dealer_score > self.player_score:
            print("Dealer wins!")
        elif self.dealer_score == self.player_score:
            print("Push")
        else:
            print("You won!")
        self.reset()
my_game = Blackjack()

def playgame():
    print('Welcome to my Blackjack game!')
    while True:
        action = input("To play type 'deal', then 'hit' or 'stand'. (to exit enter 'x' or 'c' to clear)")
        if action.lower() == 'deal':
            my_game.deal_player()
        elif action.lower() == 'hit':
            my_game.hit()
        elif action.lower() == 'stand':
            my_game.stand()
        elif action.lower() == 'x':
            return
        elif action.lower() == 'c':
            my_game.clear()
        else:
            print("You did not enter a valid command.")
playgame()
```
